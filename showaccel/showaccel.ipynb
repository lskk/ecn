```python
import json

fn = 'D:/project_amanah/S3/quakezone/M 6.9 - 102km WSW of Tugu Hilir, Indonesia/ecn.accel_2019080212_10.json'
print('Loading %s ...' % fn)
with open(fn, 'r') as f:
    accel = json.load(f)
```

```python
import numpy as np
```

```python
len(accel['n'][0])
```

```python
import datetime

accel_z = np.concatenate(accel['z'][(3*60 +27):(5*60 +30)])
accel_n = np.concatenate(accel['n'][(3*60 +27):(5*60 +30)])
accel_e = np.concatenate(accel['e'][(3*60 +27):(5*60 +30)])
x = np.arange(len(accel_z))
start_time = datetime.datetime.fromisoformat('2019-08-02T19:03:27+07:00')
print(start_time)
x_datetime = list([start_time + datetime.timedelta(milliseconds=int(25 * i)) for i in x])
```

```python
# Offset correction
accel_n = accel_n - 0.02
accel_e = accel_e - 0.3
```

```python
%matplotlib inline
#sns.lineplot(data=accel_n)
plt.figure(figsize=(25,5))
plt.plot(x_datetime, accel_z, linewidth=1, color='C0', alpha=0.7, label='Up-Down')
plt.plot(x_datetime, accel_n, linewidth=1, color='C1', alpha=0.7, label='North-South')
plt.plot(x_datetime, accel_e, linewidth=1, color='C2', alpha=0.7, label='East-West')
plt.legend()
```
